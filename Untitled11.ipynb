# Model Building and Trainning

import required model for train our model and cheak which model is better fit for my model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
```

now we are doing following steps:
1. separate test and train part form my input data
2. create a losistic regression model
3. pass my data through it
4. investigate train accuracy rate

```python
x1_train, x1_test, y1_train, y1_test = train_test_split(X1_train, Y1_train, test_size=0.3)


lr = LogisticRegression()
lr = lr.fit(x1_train,y1_train)
act = accuracy_score(y1_train,lr.predict(x1_train))
print('Train Accuracy is: ',(act*100))
```

then check test accurary

```python
x1_train, x1_test, y1_train, y1_test = train_test_split(X1_train, Y1_train, test_size=0.3)


lr = LogisticRegression()
lr = lr.fit(x1_train,y1_train)
act = accuracy_score(y1_test,lr.predict(x1_test))
print('Test Accuracy is: ',(act*100))
```

**Random Forest Model** 

```python
x1_train, x1_test, y1_train, y1_test = train_test_split(X1_train, Y1_train, test_size=0.3)
#random forest classifier with n_estimators=10 (default)
clf_rf = RandomForestClassifier(criterion = "gini", 
                                       min_samples_leaf = 3, 
                                       min_samples_split = 10,   
                                       n_estimators=100, 
                                       max_features=0.5, 
                                       oob_score=True, 
                                       random_state=1, 
                                       n_jobs=-1)
clr_rf = clf_rf.fit(x1_train,y1_train)
act = accuracy_score(y1_train,clf_rf.predict(x1_train))
print('Train Accuracy is: ',(act*100))

```

```python
x1_train, x1_test, y1_train, y1_test = train_test_split(X1_train, Y1_train, test_size=0.3)
#random forest classifier with n_estimators=10 (default)
clf_rf = RandomForestClassifier(criterion = "gini", 
                                       min_samples_leaf = 3, 
                                       min_samples_split = 10,   
                                       n_estimators=100, 
                                       max_features=0.5, 
                                       oob_score=True, 
                                       random_state=1, 
                                       n_jobs=-1)
clr_rf = clf_rf.fit(x1_train,y1_train)
act = accuracy_score(y1_test,clf_rf.predict(x1_test))
print('Test Accuracy is: ',(act*100))
```

**Decision Tree Model**
